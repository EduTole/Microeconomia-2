Información de equilibrio general puro de dos tipos de bienes (A,B)

Caso 1:

Considere dos consumidores con una función de utilidad Cobb-Douglas denotado como:

$$
u^{i}(x^{i}_{1},x^{i}_{2})=x^{i}_{1}x^{i}_{1}
$$

Para cada consumidor $i$.

Se sabe que las dotaciones son denotados como:

- Consumidor A

 $$
 (x^{A}_{1},x^{A}_{2}) =(100,350)
 $$

- Consumidor B

 $$
 (x^{B}_{1},x^{B}_{2}) =(100,50)
 $$
 
 Se le pide:
  - Curva de contratos
  - Encontrar las canastas optimas de cada consumidor (A y B) 


```python
from sympy import *
x,y = symbols('X,Y')
```

```python
Px, Py, xA, xB, yA, yB = symbols('Px, Py, xA, xB, yA, yB', positive=true)
```

Definición de la funcion de utilidad (Cobb -Douglas)

```python
UA=(xA*yA)
UB=(xB*yB)
```

Dotaciones de los bienes A y B

```python
WxA, WxB=100, 100
WyA, WyB=350, 50
```

Definicion de ingresos y gastos

```python
IncomeA =Px*WxA + Py*WyA
IncomeB =Px*WxB + Py*WyB

ExpenditureA =Px*xA+Py*yA
ExpenditureB =Px*xB+Py*yB
```

```python
MRSA =Eq(diff(UA,xA)/diff(UA,yA), Px/Py)
MRSB =Eq(diff(UB,xB)/diff(UB,yB), Px/Py)
```

```python
xSupply =WxA+WxB
ySupply =WyA+WyB
```

```python
xDemand = xA+xB
yDemand = yA+yB
```

```python
BudgetA =Eq(IncomeA, ExpenditureA)
BudgetB =Eq(IncomeB, ExpenditureB)

xMarket = Eq(xSupply, xDemand)
yMarket = Eq(ySupply, yDemand)
```

```python
solve((MRSA, MRSB, BudgetA, BudgetB, xMarket, yMarket))
```
